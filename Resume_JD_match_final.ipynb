```python
import numpy as np
import pdfplumber
import docx
from bs4 import BeautifulSoup
# from google.colab import files
# from sentence_transformers import SentenceTransformer, util
import pandas as pd
import spacy
import re
import nltk
from nltk import ngrams
from nltk.tokenize import word_tokenize

nltk.download('stopwords')
nltk.download('punkt')

# Load spaCy model for sentence segmentation
nlp = spacy.load('en_core_web_sm')

# Function to upload files in Google Colab
def upload_files():
    uploaded = files.upload()
    file_names = list(uploaded.keys())
    return file_names

# Function to extract text from PDFs
def extract_text_from_pdf(file_path):
    text=""
    with pdfplumber.open(file_path) as pdf:
        for page_num, page in enumerate(pdf.pages):
            text = page.extract_text()
    return text

# Function to extract text from DOCX files
def extract_text_from_docx(file_path):
    doc = docx.Document(file_path)
    text = ''
    for paragraph in doc.paragraphs:
        text += paragraph.text + '\n'
    return text

# Function to extract text from TXT files
def extract_text_from_txt(file_path):
    with open(file_path, 'r') as file:
        text = file.read()
    return text

# Function to preprocess text (clean HTML tags, etc.)
def preprocess_text(text):
    # Convert to lowercase
    text = text.lower()
    # Regex Cleaning
    text = re.sub(r"[^a-zA-Z0-9\s'-.]", '', text)
    # Clean html tags.
    soup = BeautifulSoup(text, 'html.parser')
    clean_text = soup.get_text()
    clean_text = clean_text.replace('\n', ' ').replace('\r', '')
    return clean_text

# Function to extract text based on file extension
def extract_text(file_path):
    if file_path.endswith('.pdf'):
        return extract_text_from_pdf(file_path)
    elif file_path.endswith('.docx'):
        return extract_text_from_docx(file_path)
    elif file_path.endswith('.txt'):
        return extract_text_from_txt(file_path)
    else:
        raise ValueError("Unsupported file format")

# Function to split text into sentences using spaCy
def split_into_sentences_spacy(text):
    doc = nlp(text)
    return [sent.text for sent in doc.sents]
```

```python
# CrossEncoding part continued...
def print_highest_similarity_scores(similarity_df, num_top=5):
    flat_index = np.unravel_index(np.argsort(-similarity_df.values, axis=None), similarity_df.shape)
    top_indices = list(zip(flat_index[0], flat_index[1]))[:num_top]

    print("\nTop Similarity Scores:")
    for i, (jd_idx, resume_idx) in enumerate(top_indices):
        jd_sentence = flat_job_description_sentences[jd_idx]
        resume_sentence = flat_resume_sentences[resume_idx]
        score = similarity_df.iloc[jd_idx, resume_idx]
        print(f"Rank {i+1}:")
        print(f"Job Description Sentence: {jd_sentence}")
        print(f"Resume Sentence: {resume_sentence}")
        print(f"Similarity Score: {score:.4f}\n")

# Print highest similarity scores
print_highest_similarity_scores(similarity_df, num_top=5)
```

```python
# CrossEncoding part continued...
# Function to find and print sentences with the highest and lowest average similarity scores
def print_highest_lowest_average_scores(similarity_df, resume_sentences, num_top=5):
    # Calculate average similarity scores for each resume sentence
    average_scores = similarity_df.mean(axis=0)

    # Get the indices of the highest and lowest average scores
    top_indices = average_scores.nlargest(num_top).index
    bottom_indices = average_scores.nsmallest(num_top).index

    print("\nTop Resume Sentences with Highest Average Similarity Scores:")
    for idx in top_indices:
        sentence_index = int(idx.split()[2]) - 1
        print(f"Resume Sentence {sentence_index + 1}: {resume_sentences[sentence_index]}")
        print(f"Average Similarity Score: {average_scores[idx]:.4f}\n")

    print("\nResume Sentences with Lowest Average Similarity Scores:")
    for idx in bottom_indices:
        sentence_index = int(idx.split()[2]) - 1
        print(f"Resume Sentence {sentence_index + 1}: {resume_sentences[sentence_index]}")
        print(f"Average Similarity Score: {average_scores[idx]:.4f}\n")

# Print highest and lowest average similarity scores
print_highest_lowest_average_scores(similarity_df, flat_resume_sentences, num_top=5)
```

```python
# METHOD 4: Dictionary method

from nltk.corpus import stopwords

# Function to preprocess text in Dictionary method.
def preprocess_text_dic(text):
    # Convert to lowercase
    text = text.lower()
    # Regex filter
    text = re.sub(r"[^a-zA-Z0-9\s'-.]", '', text)
    # Tokenize
    tokens = word_tokenize(text)
    # Remove stopwords
    stop_words = set(stopwords.words('english'))
    tokens = [word for word in tokens if word not in stop_words]
    return tokens


# Extract, preprocess text.
resume_tokens = [preprocess_text_dic(extract_text(file_name)) for file_name in resume_file_names]
job_desc_tokens = [preprocess_text_dic(extract_text(file_name)) for file_name in job_description_file_names]

print("Resume Tokens:", resume_tokens[0])
print("\n")
print("Job Description Tokens:", job_desc_tokens[0])
print("\n")
```

```python
#section 2: Identify skills. Dic() method

# Upload skill keylist.
print("Upload skill keylist (.csv)")
uploaded = files.upload()

# Get the filename
filename = list(uploaded.keys())[0]

print("Filename is: ",filename)

# Read the CSV file
skills_df = pd.read_csv(filename, header=None)
skills_df = skills_df.transpose()

skills_list = skills_df[0].tolist()

print("Skill List: ",skills_list)
```

```python
# Function to extract skills
def extract_skills(tokens, skills_list):
    return [token for token in tokens if token in skills_list]

resume_skills = set(extract_skills(resume_tokens[0], skills_list))
job_desc_skills = set(extract_skills(job_desc_tokens[0], skills_list))

print("Extracted Resume Skills:", resume_skills, "\n\n")
print("Extracted Job Description Skills:", job_desc_skills, "\n\n")
```

```python
# multi-token improvisation using n-grams.

def generate_ngrams(tokens, n):
    return list(ngrams(tokens, n))


def get_all_ngrams(tokens, max_n):
    all_ngrams = []
    for n in range(1, max_n + 1):
        all_ngrams.extend(generate_ngrams(tokens, n))
    return all_ngrams


def match_keywords(tokens, keywords, max_n=3):
    all_ngrams = get_all_ngrams(tokens, max_n)
    matches = []
    for ngram in all_ngrams:
        phrase = ' '.join(ngram)
        if phrase in keywords:
            matches.append(phrase)
    return matches


resume_matches = set(match_keywords(resume_tokens[0], skills_list, max_n=3))
job_desc_matches = set(match_keywords(job_desc_tokens[0], skills_list, max_n=3))

print("\n\nResume matches:", resume_matches, "\n\nJob Desc matches:", job_desc_matches, "\n\n")
```

```python
#section final: Calculate matching percentage. Dic() method.

# Function to calculate matching percentage
def calculate_matching_percentage(resume_skills, job_desc_skills):
    match_count = len(set(resume_skills) & set(job_desc_skills))
    total_skills = len(set(job_desc_skills))
    if total_skills == 0:
        return 0
    return (match_count / total_skills) * 100

matching_percentage = calculate_matching_percentage(resume_skills, job_desc_skills)
print("Matching Percentage single token:", matching_percentage, "\n\n")

matching_percentage = calculate_matching_percentage(resume_matches, job_desc_matches)
print("Matching Percentage multi token:", matching_percentage, "\n\n")
```

```python
# Normalized Formula for calculating final STS score. (scaled upto 20)

# 1. Mean Similarity Score using CrossEncoder. (scaled upto 10)
# 2. Matching percentage of multi-token using Dictionary method. (scaled upto 10)

normalized_mean_similarity_score = (mean_similarity*2)  # Assuming the model scores between 0 and 5.

normalized_matching_percentage = (matching_percentage/10)

# Calculate the combined score (CS) out of 20
combined_score = normalized_mean_similarity_score + normalized_matching_percentage

print(f"Mean Similarity Score (normalized to 10): {normalized_mean_similarity_score:.2f}\n\n")
print(f"Multi Token Keyword Matching Score (out of 10): {normalized_matching_percentage:.2f}\n\n")
print(f"Combined Score (out of 20): {combined_score:.2f}")
```
